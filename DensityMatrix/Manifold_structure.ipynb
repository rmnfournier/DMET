# Analyse the structure of the manifold 

## Import the data

```python
import pandas as pd
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib
import matplotlib.pyplot as plt
data = pd.read_csv("../PN_representability/data_L6_N6_super_sym/data.csv",nrows=10000,header=None,index_col=False)
print(data.shape)
```

## Dimensionality reduction
We start by trying a simple pca approach



```python
from sklearn.decomposition import PCA

pca = PCA(n_components=100)
pca.fit(data)
print(np.cumsum( pca.explained_variance_ratio_))
with PdfPages("/home/romain/Documents/EPFL/C3MP/DRLDMET/DensityMatrix/PN_representability_PCA_data_L6_N6.pdf") as export_pdf:
    matplotlib.rcParams['pdf.fonttype'] = 42
    matplotlib.rcParams['ps.fonttype'] = 42
    figure = plt.figure()
    plt.rcParams["font.family"] = "Times New Roman"
    plt.plot(pca.explained_variance_ratio_)
    plt.xlabel("Principal Components")
    plt.ylabel("EVR")
    figure.show()
    export_pdf.savefig()
    
```
